This notebook describes how to setup your machine to do development for the community site. This includes installing python, node version manager, and the required node modules. If you encounter an error in a particular section, please let us know!

For this guide, we assume you are on a Ubuntu derivative.

Make sure you are in the proper directory; since this document resides in the doc/ subfolder of the community site, we will go up one level

```bash
cd .. && pwd
```

## Installing Python

To help the developers maintain multiple environments, as we maintain a number of different projects, we utilize the python-virtualenv framework. This tool creates a "virtual python installation" for each project, thus allowing us to use Django 1.8 in one app, and Django 1.10 in another. This can be very helpfull, as you will see soon.

First, install python-virtualenv if you don't already have it

```bash
sudo apt install python-virtualenv
```

After it is installed, you will need to "activate" the virtual environment. This will download all the python dependencies.

```bash
. ./activate
```

Once all the dependicies have successfully installed, we can test the installation. By default, we do development work on a sqlite database to minimize setup time, so we don't need to configure a database, but we do need to migrate the tables to make sure everything is where it belongs.

Generally, if this command completes, you are in good shape.

```bash
python manage.py migrate
```

## Installing the Node Version Manager and Node modules

Much like how we used python-virtualenv to manage different copies of Python, most developers also use NVM to manage multiple copies of Node. Although this is strictly not required, it isn't a bad idea to utilize. Please see the NVM homepage for installation instructions (https://github.com/creationix/nvm)

```bash
curl -o- https://raw.githubusercontent.com/creationix/nvm/v0.32.1/install.sh | bash
```

Once this command completes, resart your terminal as suggested. After doing so, return to the folder that contains manage.py, and install the less compiler (lessc) which is used to compile assets.

```bash
nvm use 5
npm install -g less
```

## Testing & Launching the site

The last step is to test and launch the site. The testing isn't strictly required, but helps identify issues very quickly.

```bash
python manage.py test
```

```bash
python manage.py runserver
```

(don't hit CONTROL-C until you are done!)

Simply goto http://127.0.0.1:8000/ to access the site! Now you are ready to get some work done. Cheers!
